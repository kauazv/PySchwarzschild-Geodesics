# Simulação de Geodésicas em Schwarzschild

Este notebook demonstra a simulação numérica de geodésicas no espaço-tempo de Schwarzschild utilizando Python.

O objetivo é visualizar:

- órbitas de partículas massivas
- queda radial
- geodésicas de fótons
- desvio gravitacional da luz

O modelo é baseado na métrica de Schwarzschild:

ds² = -(1 - 2M/r) dt² + (1 - 2M/r)⁻¹ dr² + r²(dθ² + sin²θ dφ²)

```python
import sys
from pathlib import Path
sys.path.append(str(Path().resolve().parents[0]))

import numpy as np
import matplotlib.pyplot as plt
```

```python
from geodesics_massive import integrate_geodesic

M = 1.0
E = 0.95
L = 3.8

y0 = [0.0, 10.0, 0.0, 0.0]
tau_span = (0.0, 300.0)

sol = integrate_geodesic(tau_span, y0, M, E, L)

r = sol.y[1]
phi = sol.y[2]

x = r * np.cos(phi)
y = r * np.sin(phi)

plt.figure()
plt.plot(x,y)
plt.scatter([0],[0])
plt.gca().set_aspect("equal")
plt.title("Órbita relativística")
plt.show()
```
