# Phase 1 — Órbitas em Schwarzschild

Neste notebook exploramos órbitas de partículas massivas
no espaço-tempo descrito pela métrica de Schwarzschild.

Simulamos geodésicas utilizando integração numérica.

```python
import numpy as np
import matplotlib.pyplot as plt

from geodesics_massive import integrate_geodesic
```

```python
M = 1.0
E = 0.95
L = 3.8

y0 = [0.0, 10.0, 0.0, 0.0]

tau_span = (0.0, 200.0)
```

```python
solution = integrate_geodesic(
    tau_span,
    y0,
    M,
    E,
    L
)
```

```python
r = solution.y[1]
phi = solution.y[2]

x = r * np.cos(phi)
y = r * np.sin(phi)
```

```python
plt.figure(figsize=(6,6))

plt.plot(x,y)

plt.scatter([0],[0])

plt.gca().set_aspect("equal")

plt.title("Órbita relativística em Schwarzschild")

plt.show()
```
